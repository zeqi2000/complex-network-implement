```python
# BA无尺度网络模型
G = nx.random_graphs.barabasi_albert_graph(20, 1)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels = False, node_size = 30)
plt.show()

# -----------------------
# 计算获取网络每个节点的度
d = dict(nx.degree(G))
print(d)
# 计算平均度=总度数/结点数
d_avg = sum(d.values()) / len(G.nodes)  # 10.026
 
# 获取所有的度的值，及其对应的概率
# x记录有哪些度值
x = list(range(max(d.values())+1))
# 获取每个度值出现的次数
d_list = nx.degree_histogram(G)
# y计算每个度值对应的出现概率=每个度值对应的结点个数/总节点数
y = np.array(d_list) / 20
 
# 绘制度分布
plt.plot(x, y, 'o-')
plt.xlabel('du_num')
plt.ylabel('du_prob')
plt.show()
```

```python
import copy

def gene_uniform_network(t):
    G = nx.Graph()

    def add_new_edges(t):
        if t == 0:
            new_nodes = [0, 1]
            new_edges = [[0, 1]]
            G.add_nodes_from(new_nodes)
            G.add_edges_from(new_edges)
        #             return new_edges
        else:
            add_new_edges(t - 1)
            new_edges = []
            old_nodes = copy.deepcopy(G.nodes())
            for v in old_nodes:
                new_node = len(G.nodes())
                G.add_node(new_node)
                new_edges.append([v, new_node])
            G.add_edges_from(new_edges)

    add_new_edges(t)
    return G
G = gene_uniform_network(2)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300)
plt.show()
```

```python
# 自相似
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import copy

def gene_simi_network(q,t):
    G = nx.Graph()

    def add_new_edges(t):
        if t == 0:
            new_nodes = [0, 1]
            new_edges = [[0, 1]]
            G.add_nodes_from(new_nodes)
            G.add_edges_from(new_edges)
        #             return new_edges
        else:
            add_new_edges(t - 1)
            new_edges = []
            old_nodes = copy.deepcopy(G.nodes())
            for v,u in list(G.edges()):
                G.remove_edge(v,u)
                for i in range(q):
                    new_node = len(G.nodes())
                    G.add_node(new_node)
                    new_edges.append([v, new_node])
                    new_edges.append([u, new_node])

            G.add_edges_from(new_edges)

    add_new_edges(t)
    return G

G = gene_simi_network(3,2)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300)
plt.show()
```
