```python
import copy

def gene_uniform_network(t):
    G = nx.Graph()

    def add_new_edges(t):
        if t == 0:
            new_nodes = [0, 1]
            new_edges = [[0, 1]]
            G.add_nodes_from(new_nodes)
            G.add_edges_from(new_edges)
        #             return new_edges
        else:
            add_new_edges(t - 1)
            new_edges = []
            old_nodes = copy.deepcopy(G.nodes())
            for v in old_nodes:
                new_node = len(G.nodes())
                G.add_node(new_node)
                new_edges.append([v, new_node])
            G.add_edges_from(new_edges)

    add_new_edges(t)
    return G
G = gene_uniform_network(2)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300)
plt.show()
```

```python
# 
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import copy

def gene_simi_network(q,t):
    G = nx.Graph()

    def add_new_edges(t):
        if t == 0:
            new_nodes = [0, 1]
            new_edges = [[0, 1]]
            G.add_nodes_from(new_nodes)
            G.add_edges_from(new_edges)
        #             return new_edges
        else:
            add_new_edges(t - 1)
            new_edges = []
            old_nodes = copy.deepcopy(G.nodes())
            for v,u in list(G.edges()):
                G.remove_edge(v,u)
                for i in range(q):
                    new_node = len(G.nodes())
                    G.add_node(new_node)
                    new_edges.append([v, new_node])
                    new_edges.append([u, new_node])

            G.add_edges_from(new_edges)

    add_new_edges(t)
    return G

G = gene_simi_network(3,2)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300)
plt.show()
```
